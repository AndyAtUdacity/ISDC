```python
from copy import deepcopy
import simulate as sim
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

T = 250

R = 'r'
G = 'g'
grid = [
    [R,G,G,G],
    [G,G,R,G],
    [G,R,G,G],
    [R,R,G,R],
]
blur = 0.05
p_hit = 4.0
simulation = sim.Simulation(grid, blur, p_hit)

BELIEFS = []
TRUTH = []
for i in range(T):
    simulation.run()
    beliefs = simulation.beliefs
    height = len(beliefs)
    X =  []
    Y = []
    P = []
    for y, row in enumerate(beliefs):
        for x, belief in enumerate(row):
            X.append(x)
            Y.append(height-y-1) # puts large y ABOVE small y
            P.append(1000.0 * belief)
    BELIEFS.append([X,Y,P])
    TRUTH.append(simulation.true_pose)
#     BELIEFS.append([deepcopy(simulation.X), deepcopy(simulation.Y), deepcopy(simulation.P)])
#     simulation.show_beliefs()
#     time.sleep(0.3)
```
